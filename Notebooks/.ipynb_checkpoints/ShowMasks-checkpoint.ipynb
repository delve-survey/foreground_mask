```python
import healpy as hp
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../Runner/')
from AllSkyRunner import AllSkyRunner
```

# Load map and show individual masks

```python
mask = hp.read_map('/scratch/midway2/dhayaa/Foreground_Mask_Allsky.fits', dtype = int)
mask
```

```python
bits = AllSkyRunner().bits.items()

for bit, data in bits:
    mask_subset = (mask & bit) != 0
    hp.mollview(mask_subset, cbar = False)
    plt.title(data['name'] + ' (frac masked = %0.3f)' % np.average(mask_subset), size = 20)
    plt.show()
```

# Now combined mask

```python
mask_subset = (mask) != 0
hp.mollview(mask_subset, cbar = False)
plt.title('All masks (frac masked = %0.3f)' % np.average(mask_subset), size = 20)
plt.show()
```
